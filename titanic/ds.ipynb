```python
import pandas as pd
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_sur = pd.read_csv('gender_submission.csv')
df_test = df_test.merge(df_sur)
```

```python
df_train.drop(['PassengerId', 'Cabin', 'Name', 'Ticket'], axis=1, inplace=True)
df_train.head(20)
```

```python
df_train['Sex'].replace(['male', 'female'], [0, 1], inplace=True)
df_train.drop(['Embarked'], axis=1, inplace=True)
```

```python
df_train['Age'].fillna(df_train['Age'].mean(), inplace=True)
df_train.corr()
```

```python
df_train.drop(['Pclass', 'Parch'], axis=1, inplace=True)
df_train.head()
```

```python
df_train.info()
df_train.to_csv('clean_train.csv')
```

```python
df_test.drop(['Pclass', 'Parch', 'Cabin', 'Name', 'Ticket', 'Embarked', 'PassengerId'], axis=1, inplace=True)
df_test['Sex'].replace(['male', 'female'], [0, 1], inplace=True)
df_test['Age'].fillna(df_train['Age'].mean(), inplace=True)
df_test.head()
```

```python
df_test.info()
df_test.to_csv('clean_test.csv')
```
