```python
import pandas as pd
df_train = pd.read_csv('train.csv')
df_train.head(20)
```

```python
for el in df_train:
    print(el, df_train[el].unique())
```

```python
df_train.drop(['PassengerId', 'Name', 'Cabin', 'HomePlanet', 'Destination'], axis=1, inplace=True)
df_train.info()
```

```python
df_train['VIP'] = df_train['VIP'].map(
    lambda x: 1 if x == 'Yes' else 0
)
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].mean())
df_train['RoomService'] = df_train['RoomService'].fillna(df_train['RoomService'].mean())
df_train['FoodCourt'] = df_train['FoodCourt'].fillna(df_train['FoodCourt'].mean())
df_train['ShoppingMall'] = df_train['ShoppingMall'].fillna(df_train['ShoppingMall'].mean())
df_train['Spa'] = df_train['Spa'].fillna(df_train['Spa'].mean())
df_train['VRDeck'] = df_train['VRDeck'].fillna(df_train['VRDeck'].mean())

df_train['Age'] = (df_train['Age'] - df_train['Age'].mean()) / df_train['Age'].std()
df_train['RoomService'] = (df_train['RoomService'] - df_train['RoomService'].mean()) / df_train['RoomService'].std()
df_train['FoodCourt'] = (df_train['FoodCourt'] - df_train['FoodCourt'].mean()) / df_train['FoodCourt'].std()
df_train['ShoppingMall'] = (df_train['ShoppingMall'] - df_train['ShoppingMall'].mean()) / df_train['ShoppingMall'].std()
df_train['Spa'] = (df_train['Spa'] - df_train['Spa'].mean()) / df_train['Spa'].std()
df_train['VRDeck'] = (df_train['VRDeck'] - df_train['VRDeck'].mean()) / df_train['VRDeck'].std()
df_train.info()
```

```python
df_train['CryoSleep'] = df_train['CryoSleep'].map(
    lambda x: 1 if x == True else 0
)
df_train.to_csv('train_clean.csv', index=False)

df_test = pd.read_csv('test.csv')
df_test.drop(['Name', 'Cabin', 'HomePlanet', 'Destination'], axis=1, inplace=True)
df_test['VIP'] = df_test['VIP'].map(
    lambda x: 1 if x == 'Yes' else 0
)
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mean())
df_test['RoomService'] = df_test['RoomService'].fillna(df_test['RoomService'].mean())
df_test['FoodCourt'] = df_test['FoodCourt'].fillna(df_test['FoodCourt'].mean())
df_test['ShoppingMall'] = df_test['ShoppingMall'].fillna(df_test['ShoppingMall'].mean())
df_test['Spa'] = df_test['Spa'].fillna(df_test['Spa'].mean())
df_test['VRDeck'] = df_test['VRDeck'].fillna(df_test['VRDeck'].mean())
df_test['CryoSleep'] = df_test['CryoSleep'].map(
    lambda x: 1 if x == True else 0
)

df_test['Age'] = (df_test['Age'] - df_test['Age'].mean()) / df_test['Age'].std()
df_test['RoomService'] = (df_test['RoomService'] - df_test['RoomService'].mean()) / df_test['RoomService'].std()
df_test['FoodCourt'] = (df_test['FoodCourt'] - df_test['FoodCourt'].mean()) / df_test['FoodCourt'].std()
df_test['ShoppingMall'] = (df_test['ShoppingMall'] - df_test['ShoppingMall'].mean()) / df_test['ShoppingMall'].std()
df_test['Spa'] = (df_test['Spa'] - df_test['Spa'].mean()) / df_test['Spa'].std()
df_test['VRDeck'] = (df_test['VRDeck'] - df_test['VRDeck'].mean()) / df_test['VRDeck'].std()
df_test.info()
df_test.to_csv('test_clean.csv', index=False)
```
